```python
# Verification Script


import tensorflow as tf
import numpy as np
from tensorflow.keras import layers, models

# Check TensorFlow GPU access
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
print("Physical Devices: ", tf.config.list_physical_devices('GPU'))

# Define a simple CNN model for testing
def simple_cnn_model(input_shape=(32, 32, 3), num_classes=10):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(num_classes, activation='softmax'))
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Create synthetic data for testing
num_samples = 100
input_shape = (32, 32, 3)
num_classes = 10

X_train = np.random.rand(num_samples, *input_shape)
y_train = np.random.randint(0, num_classes, num_samples)

# Define and compile the model
model = simple_cnn_model(input_shape=input_shape, num_classes=num_classes)
model.summary()

# Train the model on the synthetic data
model.fit(X_train, y_train, epochs=5, batch_size=10)

# Evaluate the model
loss, accuracy = model.evaluate(X_train, y_train)
print(f"Test loss: {loss}")
print(f"Test accuracy: {accuracy}")
```

```python
import os

# Function to execute a command and capture the output
def run_command(command):
    stream = os.popen(command)
    output = stream.read()
    return output

# Get NVIDIA driver version
nvidia_driver_version = run_command("nvidia-smi")

# Get CUDA version
cuda_version = run_command("nvcc --version")

# Get CUDA toolkit version
cuda_toolkit_version = run_command("cat /usr/local/cuda/version.txt")

# Get cuDNN version
cudnn_version = run_command("cat /usr/include/cudnn_version.h | grep CUDNN_MAJOR -A 2")

# Get GPU info
gpu_info = run_command("lspci | grep -i nvidia")

# Get environment variables
environment_variables = run_command("printenv | grep CUDA")

# Display the gathered information
print("NVIDIA Driver Version:")
print(nvidia_driver_version)
print("CUDA Version:")
print(cuda_version)
print("CUDA Toolkit Version:")
print(cuda_toolkit_version)
print("cuDNN Version:")
print(cudnn_version)
print("GPU Info:")
print(gpu_info)
print("Environment Variables:")
print(environment_variables)
```

```python
import os

# Function to execute a command and capture the output
def run_command(command):
    stream = os.popen(command)
    output = stream.read()
    return output

# Get NVIDIA driver version
nvidia_driver_version = run_command("nvidia-smi")

# Get CUDA version
cuda_version = run_command("nvcc --version")

# Get CUDA toolkit version
cuda_toolkit_version = run_command("cat /usr/local/cuda/version.txt")

# Get cuDNN version
cudnn_version = run_command("cat /usr/include/cudnn_version.h | grep CUDNN_MAJOR -A 2")

# Get GPU info
gpu_info = run_command("lspci | grep -i nvidia")

# Get environment variables related to CUDA
cuda_environment_variables = run_command("printenv | grep CUDA")

# Get environment variables related to NVIDIA
nvidia_environment_variables = run_command("printenv | grep NVIDIA")

# Get operating system details
os_details = run_command("uname -a")

# Get detailed memory info
memory_info = run_command("cat /proc/meminfo")

# Get detailed CPU info
cpu_info = run_command("cat /proc/cpuinfo")

# Get list of installed packages (Python packages)
installed_packages = run_command("pip list")

# Get list of installed system packages (for Debian-based systems)
installed_system_packages = run_command("dpkg -l")

# Display the gathered information
print("NVIDIA Driver Version:")
print(nvidia_driver_version)
print("\nCUDA Version:")
print(cuda_version)
print("\nCUDA Toolkit Version:")
print(cuda_toolkit_version)
print("\ncuDNN Version:")
print(cudnn_version)
print("\nGPU Info:")
print(gpu_info)
print("\nCUDA Environment Variables:")
print(cuda_environment_variables)
print("\nNVIDIA Environment Variables:")
print(nvidia_environment_variables)
print("\nOperating System Details:")
print(os_details)
print("\nMemory Info:")
print(memory_info)
print("\nCPU Info:")
print(cpu_info)
print("\nInstalled Python Packages:")
print(installed_packages)
print("\nInstalled System Packages:")
print(installed_system_packages)
```
